```python
#2a
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import warnings
warnings.filterwarnings("ignore")


# Load the CSV file
df = pd.read_csv('Marketing Campaign data.csv')

# Separate features and target variable
X = df.drop(columns=['AFFINITY_CARD']) #Feature
y = df['AFFINITY_CARD'] #target

# Define significance level for chi-square test
alpha = 0.05

# Remove variables with zero or weak correlation with target variable
numerical_corr = X.select_dtypes(include=['float64', 'int64']).apply(lambda x: x.corr(y)) # For Numeric

variablesToKeep = numerical_corr[abs(numerical_corr) > 0.1].index.tolist()  # Adjust correlation threshold as needed

# chi-square test for categorical variables
categorical_cols = X.select_dtypes(include=['object']).columns
for col in categorical_cols:
    contingency_table = pd.crosstab(X[col], y)
    chi2, p, _, _ = chi2_contingency(contingency_table)
    if p < alpha:
        variablesToKeep.append(col)  # Change to variablesToKeep

# Filter the DataFrame to keep only selected variables
df_filtered = df[variablesToKeep + ['CUST_INCOME_LEVEL'] + ['AFFINITY_CARD']]


df_filtered

# Save the filtered DataFrame to a new CSV file
#df_filtered.to_csv('filtered_data2.csv', index=False)
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import random

data = df10


#Select 1000 random customer records
random.seed(23)  # For reproducibility
selected_indices = random.sample(range(len(data)), 1000)
data_selected = data.iloc[selected_indices]

# Prepare the data
X_train = data_selected.drop(columns=["AFFINITY_CARD"])  # Features
y_train = data_selected["AFFINITY_CARD"]  # Target variable



# Build and train the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score

# Validation functions
def validate_age(age):
    try:
        age = int(age)
        if age <= 0:
            return False
        return True
    except ValueError:
        return False

def validate_years_of_residence(years):
    try:
        years = int(years)
        if years < 0 or years > 14:  # Add the range of acceptable values
            return False
        return True
    except ValueError:
        return False
    
def validate_theater(theater):
    try:
        theater = int(theater)
        if theater == 0 or theater==1:  # Add the range of acceptable values
            return True
        return False
    except ValueError:
        return False    
    
def validate_bookkeeping(bookkeeping):
    try:
        bookkeeping = int(bookkeeping)
        if bookkeeping == 0 or bookkeeping==1:  # Add the range of acceptable values
            return True
        return False
    except ValueError:
        return False    
    
def validate_ybox(ybox):
    try:
        ybox = int(ybox)
        if ybox == 0 or ybox==1:  # Add the range of acceptable values
            return True
        return False
    except ValueError:
        return False    
    
def validate_gender(gender):
    try:
        gender = int(gender)
        if gender == 0 or gender==1:  # Add the range of acceptable values
            return True
        return False
    except ValueError:
        return False    
    
def validate_meritalstatus(meritalstatus):
    try:
        meritalstatus = int(meritalstatus)
        if meritalstatus == 0 or meritalstatus==1 or meritalstatus==2 or meritalstatus==3 or meritalstatus==4 or meritalstatus==5 or meritalstatus==6:  # Add the range of acceptable values
            return True
        return False
    except ValueError:
        return False    
    
def validate_country(country):
    try:
        country = int(country)
        if country == 0 or country==1 or country==2 or country==3 or country==4 or country==5 or country==6 or country==7 or country==8 or country==9 or country==10 or country==11 or country==12 or country==13 or country==14 or country==15 or country==16 or country==17 or country==18 or country==19:  # Add the range of acceptable values
            return True
        return False
    except ValueError:
        return False    
    
def validate_education(education):
    try:
        education = int(education)
        if education == 0 or education==1 or education==2 or education==3 or education==4 or education==5 or education==6 or education==7 or education==8:  # Add the range of acceptable values
            return True
        return False
    except ValueError:
        return False    
    

def validate_occupation(occupation):
    try:
        occupation = int(occupation)
        if 0 <= occupation <= 14 :  # Add the range of acceptable values
            return True
        return False
    except ValueError:
        return False    
    
def validate_household(household):
    try:
        household = int(household)
        if 0 <= household <= 6:  # Add the range of acceptable values
            return True
        return False
    except ValueError:
        return False    
    
def validate_income(income):
    try:
        income = int(income)
        if 1 <= income <=3:  # Add the range of acceptable values
            return True
        return False
    except ValueError:
        return False    

# Add similar validation functions for other fields

# Prediction application with user interface
def predict_customer_affinity(model, customer_data):
    """
    Predict customer affinity using the trained logistic regression model.

    Args:
    - model: Trained logistic regression model
    - customer_data: Dictionary containing customer data

    Returns:
    - Prediction: Predicted customer affinity (0 or 1)
    """
    
    # Convert customer data to DataFrame
    df = pd.DataFrame(customer_data, index=[0])
    
    # Predict customer affinity
    prediction = model.predict(df)
    
    return prediction[0]

def main():
    # User interface
    print("Welcome to the Customer Affinity Prediction Application!")
    print("Please enter customer data:")

    # Accept user input
    customer_data = {}

    # Age validation
    while True:
        age = input("Age: ")
        if validate_age(age):
            customer_data['AGE'] = int(age)
            break
        else:
            print("Please enter a valid age (a positive integer).")

    # Years of Residence validation
    while True:
        years = input("Years of Residence[0-14]: ")
        if validate_years_of_residence(years):
            customer_data['YRS_RESIDENCE'] = int(years)
            break
        else:
            print("Please enter a valid number of years of residence within the range [0-14].")

    # Similar validation for other fields


    # HOME_THEATER_PACKAGE validation
    while True:
        theater = input("Home Theater Package (0 or 1): ")
        if validate_theater(theater):
            customer_data['HOME_THEATER_PACKAGE'] = int(theater)
            break
        else:
            print("Please enter a valid number for availability of theater")


    # BOOKKEEPING_APPLICATION validation
    while True:
        bookkeeping = input("Bookkeeping Application (0 or 1): ")
        if validate_bookkeeping(bookkeeping):
            customer_data['BOOKKEEPING_APPLICATION'] = int(bookkeeping)
            break
        else:
            print("Please enter a valid number for availability of book keeping application")


    # Y_BOX_GAMES validation
    while True:
        ybox = input("Y-Box Games (0 or 1): ")
        if validate_ybox(ybox):
            customer_data['Y_BOX_GAMES'] = int(ybox)
            break
        else:
            print("Please enter a valid number for availability of Y-game box")

    # CUST_GENDER validation
    while True:
        gender = input("Customer Gender (0 for female, 1 for male ): ")
        if validate_gender(gender):
            customer_data['CUST_GENDER'] = int(gender)
            break
        else:
            print("Please enter a valid number gender(0-F or 1-M)")

    # CUST_MARITAL_STATUS validation   
    while True:
        meritalstatus = input("Customer Marital Status [0-Divorced, 1-Mabsent, 2-Mar-AF, 3-Married, 4-NeverM, 5-Separeted, 6-Widowed]: ")
        if validate_meritalstatus(meritalstatus):
            customer_data['CUST_MARITAL_STATUS'] = int(meritalstatus)
            break
        else:
            print("Please enter a valid number according to indicates merital status")


    # COUNTRY_NAME validation   
    while True:
        country = input("Country Name [ Spain-1, France-2, South Africa-3, Turkey-4, Australia-5, Japan-6, New Zealand-7, Singapore-8, China-9, Saudi Arabia-10, Denmark-11, United Kingdom-12, Poland-13, Germany-14, Canada-15, Brazil-16, Italy-17,Argentina-18, United States of America-19]: ")
        if validate_country(country):
            customer_data['COUNTRY_NAME'] = int(country)
            break
        else:
            print("Please enter a valid number belongs to the country")


    # EDUCATION validation  
    while True:
        education = input("Education Level[ 'Presch.': 1, '1st-4th': 2, '5th-6th': 3, '7th-8th': 4, '9th': 5, '10th': 5,'11th': 5,'12th': 5,'HS-grad': 5,'Profsc': 6, '< Bach.': 6,'Assoc-A': 6,'Assoc-V': 6,'Bach.': 7,'Masters': 7, 'PhD': 8]:")
        if validate_education(education):
            customer_data['EDUCATION'] = int(education)
            break
        else:
            print("Please enter a valid number according to study level")


    # OCCUPATION validation  
    while True:
        occupation = input("Occupation [Armed-F - 1,'Cleric.-2','Crafts-3','Exec.-4','Farming-5', 'Handler-6','House-s-7','Machine-8','Other-9','Prof.-10', 'Protec.-11','Sales-12','TechSup-13','Transp.-14] : ")
        if validate_occupation(occupation):
            customer_data['OCCUPATION'] = int(occupation)
            break
        else:
            print("Please enter a valid number as per occupation")


    # HOUSEHOLD_SIZE validation  
    while True:
        household = input("Household Size [1-6] as well as 6 indicates 9+: ")
        if validate_household(household):
            customer_data['HOUSEHOLD_SIZE'] = int(household)
            break
        else:
            print("Please enter a valid number as per household-size")

    # CUST_INCOME_LEVEL validation  
    while True:
        income = input("Customer Income Level[low-1, medium-2, high-3]: ")
        if validate_income(income):
            customer_data['CUST_INCOME_LEVEL'] = int(income)
            break
        else:
            print("Please enter a valid number for income level")

    
    
    # Predict customer affinity
    prediction = predict_customer_affinity(model, customer_data)
    
    # Display prediction
    print("Predicted Customer Affinity:", prediction)

if __name__ == "__main__":
    main()
```
